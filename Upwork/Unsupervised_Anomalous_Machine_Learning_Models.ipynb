### Projection based approach : **Isolation Forest**

Isolation Forest identifies anomalies or outliers rather than profiling normal data points. It isolates observations by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of that selected feature. This split depends on how long it takes to separate the points.

```python
from sklearn.ensemble import IsolationForest

#Initialize Isolation Forest
clf=IsolationForest(n_estimators=100, max_samples='auto', contamination=float(.007), \
                        max_features=1.0, bootstrap=False, n_jobs=-1, random_state=42, verbose=0)

# Fit the data
clf.fit(df_scaled.values)

# Predict clusters
pred = clf.predict(df_scaled.values)

# Append pred in space_df
space_df['cluster_iso']=pred
# Data point is anomaly if pred == -1
outliers = space_df.loc[space_df['cluster_iso']==-1]
outlier_index_iso=list(outliers.index)
outlier_index_iso
```

Now, We will try to visualize the results and check if the clustering makes sense.
Normalize and fit the dimensions to a PCA to reduce the number of dimensions and then plot them in 3D highlighting the anomalies.

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

# Define a function to plot 3D highlighting the anomalies
def Axes3d_plot(df, outlier_index, method):
  # Initialize PCA
  pca = PCA(n_components=3)  # Reduce to k=3 dimensions
  # Fit
  df_pca = pca.fit_transform(df)

  # Plot PCA components
  fig = plt.figure(figsize = (12, 6))
  ax = fig.add_subplot(111, projection='3d')
  ax.set_title(f'PCA components with normal stars, & anomalies - {method}')
  ax.set_xlabel("X component")
  ax.set_ylabel("Y component")
  ax.set_zlabel("Z component")
  # Plot the compressed data points
  ax.scatter(df_pca[:, 0], df_pca[:, 1], zs=df_pca[:, 2], s=4, lw=1, label="normal",c="green")
  # Plot x's for the ground truth outliers
  ax.scatter(df_pca[outlier_index,0],df_pca[outlier_index,1], df_pca[outlier_index,2],
           lw=2, s=60, marker="x", c="red", label="Anomolies")
  ax.legend()
  plt.show()
```

```python
# Plot 3D for Isolation forest
Axes3d_plot(df_scaled, outlier_index_iso, 'Isolation Forest')
```

### Density based approach : **Kernel Density estimator** (KDE)

 It helps us measure kernel density of samples which can be then used to take out outliers. It uses KDTree or BallTree algorithm for kernel density estimation. We are using 'gaussian' kernel for this estimation.

```python
from sklearn.neighbors.kde import KernelDensity

# Estimate density with a Gaussian kernel density estimator
kde = KernelDensity(kernel='gaussian')
kde.fit(df_scaled)
```

Calculate Log Density Evaluations for Each Sample using score_samples() function.

```python
df_kde = kde.score_samples(df_scaled)
```

**Dividing Dataset into Valid Samples and Outliers**

We'll divide these values into 99.3% as valid data and 0.7% as outliers based on the output of score_samples() function.

```python
from scipy.stats.mstats import mquantiles

alpha_set = 0.993   # Data excluding outliers
tau_kde = mquantiles(df_kde, 1. - alpha_set)       # Threshold value
tau_kde
```

All the values in **df_kde** array which are less than **tau_kde** will be outliers and values greater than it will be qualified as valid samples. We'll try to find out indexes of samples that are outliers and valid. We'll then use these indexes to filter data to divide it into outliers and valid samples.

```python
outliers = np.argwhere(df_kde < tau_kde)
outliers_index_kde = outliers.flatten()
kde_outliers = space_df.values[outliers_index_kde]

normal_samples = np.argwhere(df_kde >= tau_kde)
normal_samples = normal_samples.flatten()
kde_valid = space_df.values[normal_samples]

print("Original Samples : ",space_df.shape[0])
print("Number of Anomalies : ", len(outliers))
print("Number of Normal Samples : ", len(normal_samples))
```

```python
# Plot 3D for Isolation forest
Axes3d_plot(df_scaled, outliers_index_kde, 'KDE')
```

### Quantile based approach : **OneClassSVM**

 It's based on the SVM algorithm which is used behind the scene to make a decision about the sample is anomaly or not. We are considering '**rbf**' kernel for this method.

```python
from sklearn.svm import OneClassSVM

nu = 0.007  # percentage of outliers
ocsvm = OneClassSVM(kernel='rbf', gamma=0.002, nu=nu)
ocsvm.fit(df_scaled)
```

### Neighbor based approach : **Local Outlier Factor** (LOF)

It measures the local density of a sample with respect to its neighbors. Based on a comparison between the local density of the sample and its neighbors, the decision is made whether the sample is an outlier or not.

```python
from sklearn.neighbors import LocalOutlierFactor
# Initialize LOF
lof = LocalOutlierFactor(contamination = 0.007, metric = 'euclidean')
```

**LOF** provides **fit_predict()** method which accepts samples and returns array consisting of values 1 or -1. Here 1 represents a **valid sample** and -1 represents an **outlier**.

```python
preds = lof.fit_predict(df_scaled)
space_df['cluster_lof'] = preds
```

Dividing Dataset into Valid Samples and Outliers

```python
lof_outliers = space_df.values[preds == -1]   # -1 --> anomalies
lof_valid = space_df.values[preds != -1]
lof_outliers_index = space_df.index[preds == -1]

print("Original Samples : ",space_df.shape[0])
print("Number of Outliers : ", lof_outliers.shape[0])
print("Number of Normal Samples : ", lof_valid.shape[0])
```

**Data Analysis**

Since **Isolation Forest, KDE, & OneClassSVM** results are comparable, We will consider one method for data analysis. Let's consider **OneClassSVM** method for data analysis.

```python
# For ploting, drop stars column & convert -1 to "Anomaly", 1 to "Normal"  in cluster column
df_final = space_df.drop('stars', axis=1)
def apply(row):
  if row ==-1:
    return "Anomaly"   # Anomaly
  else:
    return "Normal"   # Normal

# Apply above function
df_final.cluster_ocsvm = df_final.cluster_ocsvm.apply(lambda x : apply(x))
```

Let's see how normal & anomoly stars are distributed.

```python
import matplotlib.pyplot as plt

import matplotlib.gridspec as gridspec

def create_figure(df, plot1_f1, plot1_f2, plot2_f):
    with plt.style.context(("seaborn","ggplot")):
        fig = plt.figure(constrained_layout=True, figsize=(10,8))
        specs = gridspec.GridSpec(ncols=1, nrows=2, figure=fig) ## Declaring 1x2 figure.

        ax1 = fig.add_subplot(specs[0, :]) ## First Row
        ax2 = fig.add_subplot(specs[1, 0]) ## Second Row

        ## First Graph -  Scatter Plot 
        color = {"Anomaly" : "tab:red", "Normal":"tab:green"}
        for i in df['cluster_ocsvm'].unique():
            ax1.scatter(df[df['cluster_ocsvm']==i][plot1_f1],
                        df[df['cluster_ocsvm']==i][plot1_f2],
                        c=color[i],
                        s=200,
                        alpha=0.6,
                        label=i)
        ax1.set_xlabel(plot1_f1)
        ax1.set_ylabel(plot1_f2)
        ax1.set_title("%s vs %s Scatter Plot"%(plot1_f1, plot1_f2))
        ax1.legend(title="Star type",loc="best")

        ## Second Graph - Histogram
        ax2.hist(df[plot2_f], color="tab:orange", )
        ax2.set_xlabel(plot2_f)
        ax2.set_ylabel('Count')
        ax2.set_title("%s distribution"%(plot2_f))

        plt.close(fig)
        return fig
```

To make these plots interactive, we have Panel library. This library makes our work quite easy to create dashboard. We'll start loading panel by importing it and then calling it's extension() method to load it. We'll then create dictionary which will have list of possible values for all parameters of create_figure() method. We can then convert our simple create_figure() function into dashboard by just calling pn.interact() method on it passing dictionary of all possible parameter values.

```python
import panel as pn
import panel.widgets as pnw
```

We'll start by creating 3 dropdowns each of which will have all features as options. We also have given names to each dropdown about which parameter it belongs to.

```python
f1 = pnw.Select(options=list(df_final.columns), name="Scatter Plot Feature 1")
f2 = pnw.Select(options=list(df_final.columns), name="Scatter Plot Feature 2")
f3 = pnw.Select(options=list(df_final.columns), name="Histogram Feature")
```

We can link the above-mentioned widgets to a parameter of our method by creating wrapper function around it and giving it decorator (@pn.depends) as mentioned below. It'll link the value of these widgets with a value of a function.

```python
@pn.depends(f1.param.value,f2.param.value,f3.param.value)
def create_dash(f1,f2,f3):
    return create_figure(df_final, f1, f2, f3)
```

```python
pn.extension()
heading = "<h3>Space Data Analysis Dashboard </h3>\nPlease feel free to try various combinations of parameters to analyze results from different angles"
text = pnw.StaticText(value=heading, align="center")

dash = pn.Column(
                    text,
                    pn.Row(f1,f2, align="center"),
                    pn.Row(f3, align="center"),
                    pn.Row(create_dash)
                )
dash
```
